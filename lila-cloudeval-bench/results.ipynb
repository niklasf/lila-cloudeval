```python
import seaborn as sns
```

```python
import pandas as pd
```

```python
data = [
    pd.read_csv(path, sep="\t", names=["fen", label], skiprows=0)[label]
    for label, path in [
        ("multi pv 1", "results-1.tsv"),
        ("multi pv 5", "results-5.tsv"),
    ]
]
```

```python
sns.set_theme(style="darkgrid", rc={
    "figure.figsize": (10, 6),
    "xtick.bottom": True,

    "figure.facecolor": "#302e2c",
    "axes.labelcolor": "#bababa",
    "xtick.color": "#bababa",
    "ytick.color": "#bababa",
})

ax = sns.ecdfplot([series + 1 for series in data], log_scale=True)
ax.set(ylim=(0.85, 1.0), xlabel="Response time in milliseconds (including 1ms offset)", ylabel="Proportion faster than response time")
sns.move_legend(ax, "lower right")
ax
```
